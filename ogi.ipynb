```python
import random
import pandas as pd
import numpy as np
import time
import matplotlib.pyplot as plt
import datetime
```

```python
data = pd.read_csv('customer_information.csv')
data.head(10)
```

```python
def generate_sample_id(index):
    return f"sid_{str(index + 1).zfill(5)}"

data['id'] = data.index.to_series().apply(generate_sample_id)

data.head()
```

```python
# Assuming 'data' is your DataFrame
number_country = data['n_countries_visited'].value_counts()

# Sort the values and counts in increasing order of the values
sorted_values_counts = number_country.sort_index()

# Display the result
import pandas as pd

# Assuming 'data' is your DataFrame
bins = list(range(0, max(data['n_countries_visited']) + 6, 5))
labels = [f"{start}-{end - 1}" for start, end in zip(bins[:-1], bins[1:])]

# Create a new column with the specified bins
data['n_countries_range'] = pd.cut(data['n_countries_visited'], bins=bins, labels=labels, include_lowest=True)

# Count the occurrences in each bin
range_counts = data['n_countries_range'].value_counts()

# Display the result
print("Values in ranges of 5 and their counts:")
print(range_counts)
```
