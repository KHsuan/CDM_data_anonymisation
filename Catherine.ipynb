```python
import random
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('customer_information.csv')
data.head(10)
```

### avg_n_drinks_per_week
banding (0-2,2-4) into low moderate heavy 

```python
data["avg_n_drinks_per_week"].describe()
```

```python
plt.hist(data["avg_n_drinks_per_week"])
plt.xlabel("Average number drinks per week")
plt.ylabel("People")
plt.show()
```

```python
bins= [0,2,4,6,8,12]
labels = ["0 to <2", "2 to <4", "4 to <6", "6 to <8", "8 to 10"]
data["avg_n_drinks_per_week_group"] = pd.cut(data['avg_n_drinks_per_week'], bins=bins, labels=labels, right=False)
```

```python
counts = data["avg_n_drinks_per_week_group"].value_counts()
count_in_order = []
for i in labels:
    count_in_order.append(counts[i])

plt.bar(labels, count_in_order)
plt.xlabel("Average number drinks per week in group")
plt.ylabel("People")
plt.show()
```

```python
bins= [0,1,4,14]
labels = ["never", "light", "moderate"]
data["avg_n_drinks_per_week_group"] = pd.cut(data['avg_n_drinks_per_week'], bins=bins, labels=labels, right=False)
```

```python
counts = data["avg_n_drinks_per_week_group"].value_counts()
count_in_order = []
for i in labels:
    count_in_order.append(counts[i])

plt.bar(labels, count_in_order)
plt.xlabel("Average number drinks per week in group")
plt.ylabel("People")
plt.show()
```
