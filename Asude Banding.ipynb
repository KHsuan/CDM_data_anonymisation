```python
import random
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('customer_information.csv')
data.head(10)
```

```python
# Banding postcode to partial postcode (first 2 characters)
data['postcode_banded'] = data['postcode'].astype(str).str[-1]

# shows both columns: print(data[['postcode', 'postcode_banded']])
# Replacing original column with banded column
data['postcode'] = data['postcode_banded']

# Drop the 'postcode_banded' column if no longer needed
data.drop(columns=['postcode_banded'], inplace=True)

# Print the updated DataFrame with the banded postcode column
print(data['postcode'])

# Find number unique postcodes and print
num_unique_postcodes = data['postcode'].nunique()
print("Number of unique postcodes:", num_unique_postcodes)
```

```python
postcode_counts = data['postcode'].value_counts()

print(postcode_counts)
```

```python
# Before banding, determining range of weight by finding minimum and maximum values
min_weight = data['weight'].min()
max_weight = data['weight'].max()

print(f"Minimum weight: {min_weight}")
print(f"Maximum weight: {max_weight}")
```

```python
bin_edges = list(range(35, 111, 10))  # Ranges from 35 to 100 in increments of 10
bin_labels = [f"{i}-{i+10}" for i in range(35, 101, 10)]  # Labels such as '35-45', '45-55', ...
data['weight_ranges'] = pd.cut(data['weight'], bins=bin_edges, labels=bin_labels, right=False) # Create bins and categorize weights

# Shows both original and new banded weight columns: print(data[['weight', 'weight_ranges']])
# Replacing weight column with weight_ranges
data['weight'] = data['weight_ranges']

# Drop the 'weight_ranges' column if no longer needed
data.drop(columns=['weight_ranges'], inplace=True)

# Print the updated DataFrame with the banded weight column
print(data['weight'])

# Find the number of unique weight ranges
num_unique_weights = data['weight'].nunique()
print("Number of unique weight ranges:", num_unique_weights)
```

```python
# checking number of unique values
unique_weight = data['weight'].value_counts()
unique_weight
```

```python
min_height = data['height'].min()
max_height = data['height'].max()

print(f"Minimum height: {min_height}")
print(f"Maximum height: {max_height}")
```

```python
# Banding height into increments of 0.2
# Define the specific ranges for height categories
height_bins = [1.4, 1.6, 1.8, 2.1]  # Ranges for categories: 1.4-1.6, 1.6-1.8, 1.8-2.0 (adjust upper bound to include 2.0)
height_labels = ['1.4-1.6', '1.6-1.8', '1.8-2.0']  # Labels for the categories

# Banding heights into specified ranges
data['height'] = pd.cut(data['height'], bins=height_bins, labels=height_labels, right=False)

# Print the updated DataFrame with the banded height column
print(data['height'])

# Find the number of unique height categories
num_unique_heights = data['height'].nunique()
print("Number of unique height categories:", num_unique_heights)
 
```
